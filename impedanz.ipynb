```python
# NUR AUSFÜHREN, WENN MAN IM NACHHINEIN IM FILE "impedanz_data.csv" ZEILE 2,3 UND 4 LÖSCHT!!!!!!!!!!!!

# Required libraries
import csv

# make handy csv file
file_path = 'impedanz_data.txt'

# Initialize an empty list to store the lines
lines_list = []

# Open the file for reading
with open(file_path, 'r') as file:
    # Read each line in the file one by one
    for line in file:
        # Strip the newline characters and add the line to the list
        lines_list.append(line.strip())

# Now lines_list contains each line of the file as a list element
# print(lines_list)

textfile_complete_list = []
for datasample in lines_list:
    # print("1:", datasample)
    # print("2", datasample.split("\t"))
    textfile_complete_list.append(datasample.split("\t"))
    
# print("compleTTT", textfile_complete_list)

# print("list1:", list1)

# Zipping the lists together to pair each value across the three strings
# paired_data = zip(list1, list2, list3)
MINE_paired_data = zip(*textfile_complete_list)

# print("paire:", MINE_paired_data)

# CSV file path where you want to save the file
csv_file_path = 'impedanz_data.csv'

# Writing to CSV
with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    #writer.writerow(["spalte1", "spalte2", "spalte3"])  # Writing the header
    writer.writerows(MINE_paired_data)  # Writing the data rows
```

```python
# Funktion, um Plots zu machen

import numpy as np 
import matplotlib.pyplot as plt 

# plot with two axes
def plot_with_two_axes(data_x,data_y_one, data_y_two, name_x, name_y1, name_y2, plot_title="title", filename_to_save="LOL.pdf", xmin=0, xmax=500000, ymin1=0, ymax1=1000, ymin2=-2, ymax2=1):
    fig, ax1 = plt.subplots() 
    
    ax1.set_xlabel(name_x) 
    ax1.set_ylabel(name_y1, color = 'red') 
    plot_1 = ax1.plot(data_x, data_y_one, color = 'red', label='impedance') 
    ax1.tick_params(axis ='y') #, labelcolor = 'red') 
    # ax1.ylim(0,.9)
    # ax1.xlim(0,.9)
    ax1.set_xlim([xmin, xmax])
    ax1.set_ylim([ymin1, ymax1])

    ax2 = ax1.twinx() 
    ax2.set_ylabel(name_y2, color = 'blue') 
    plot_2 = ax2.plot(data_x, data_y_two, color = 'blue', label = 'phase') 
    ax2.tick_params(axis ='y') #, labelcolor = 'blue') 

    ax1.set_xlim([xmin, xmax])
    ax2.set_ylim([ymin2, ymax2])
    
    fig.tight_layout()

    lns = plot_1 + plot_2
    labels = [l.get_label() for l in lns]
    
    plt.title(plot_title)
    plt.legend(lns, labels, loc=0)
    
    fnts =  filename_to_save + ".pdf"
    plt.savefig(fnts)
    plt.show()
    
    



```

```python
# load sarah data

import pandas as pd

# Path to your data file
file_path = 'R_1MR.txt'

# List to store the rows
data_naeherungen = []

# Open and read the file
with open(file_path, 'r') as file:
    for line in file:
        # Strip newlines and any trailing whitespace
        line = line.strip()
        
        # Split on the first tab to isolate the first column
        parts = line.split('\t')
        
        if len(parts) >= 2:
            # First part is column1
            column1 = float(parts[0])
            
            # Second part needs further splitting on comma for column2 and column3
            column2_and_3 = parts[1].split(',')
            
            if len(column2_and_3) == 2:
                column2 = float(column2_and_3[0])
                column3 = float(column2_and_3[1])
                
                # Append the parsed columns to the data list
                data_naeherungen.append([column1, column2, column3])

# Create a DataFrame
df = pd.DataFrame(data_naeherungen, columns=['column1', 'column2', 'column3'])

print(df.column1)


# Print the DataFrame to check if it's correct
# print(df)
```
