```python
# replace first line with more meaningful stuff

def change_first_line_sarah(filename):
    # Path to the file
    file_path = filename

    # Read the file
    with open(file_path, 'r') as file:
        lines = file.readlines()

    # Replace the first line
    lines[0] = 'frequency impedance phase\n'

    # Write the modified lines back to the file
    with open(file_path, 'w') as file:
        file.writelines(lines)
```

```python
# load sarah data

import pandas as pd

def read_sarah_file(filename):

    # Path to your data file
    file_path = filename

    # List to store the rows
    data_naeherungen = []

    # Open and read the file
    with open(file_path, 'r') as file:
        for line in file:
            # Strip newlines and any trailing whitespace
            line = line.strip()
            
            # Split on the first tab to isolate the first column
            parts = line.split('\t')
            
            if len(parts) >= 2:
                # First part is column1
                column1 = float(parts[0])
                
                # Second part needs further splitting on comma for column2 and column3
                column2_and_3 = parts[1].split(',')
                
                if len(column2_and_3) == 2:
                    column2 = float(column2_and_3[0])
                    column3 = float(column2_and_3[1])
                    
                    # Append the parsed columns to the data list
                    data_naeherungen.append([column1, column2, column3])

    # Create a DataFrame
    df = pd.DataFrame(data_naeherungen, columns=['column1', 'column2', 'column3'])

    return df
```

### Daten umrechnen in Impendaz und Rad von Imag und Real


```python
#Schreibt Real und ImaginÃ¤rteil vom Txt um --> Impedanz und phase!!


import numpy as np
import os

def berechne_impedanz_und_phase(input_datei_pfad, output_datei_pfad):
    # Liste zum Speichern der Ergebnisse
    ergebnisse = []
    
    # Die erste Zeile (Ãberschrift) zur Ausgabedatei hinzufÃ¼gen
    ergebnisse.append("Frequenz Impedanz Phase\n")

    # Datei Ã¶ffnen und Zeile fÃ¼r Zeile lesen
    with open(input_datei_pfad, 'r') as file:
        lines = file.readlines()
        
        # Erste Zeile Ã¼berspringen (Ãberschriften oder nicht benÃ¶tigte Informationen)
        for line in lines[1:]:  # Hier wird die erste Zeile Ã¼bersprungen
            # Teile die Zeile in die einzelnen Spalten
            spalten = line.strip().split()
            
            # Frequenz extrahieren (erste Spalte)
            frequenz = float(spalten[0])
            
            # Real- und ImaginÃ¤rteil (zweite Spalte, durch Komma getrennt)
            real_imag = spalten[1].split(',')
            real_teil = float(real_imag[0])
            imag_teil = float(real_imag[1])
            
            # Berechnung der Impedanz (Betrag der komplexen Zahl)
            impedanz = np.sqrt(real_teil**2 + imag_teil**2)
            
            # Berechnung der Phase in Radiant und Umrechnung in Grad
            phase_rad = np.arctan2(imag_teil, real_teil)
            phase_grad = np.degrees(phase_rad)
            
            # Ergebniszeile formatieren (detaillierte Formatierung wie in der Eingabedatei)
            ergebnisse.append(f"{frequenz:.10e} {impedanz:.10e} {phase_grad:.10f}\n")
    
    # Ergebnisse in eine neue Datei schreiben
    with open(output_datei_pfad, 'w') as out_file:
        out_file.writelines(ergebnisse)

    print(f"Ergebnisse wurden in die Datei '{output_datei_pfad}' geschrieben.")

def verarbeite_alle_dateien(input_verzeichnis, output_verzeichnis):
    # Sicherstellen, dass das Output-Verzeichnis existiert
    if not os.path.exists(output_verzeichnis):
        os.makedirs(output_verzeichnis)

    # Alle Dateien im Input-Verzeichnis auflisten
    for dateiname in os.listdir(input_verzeichnis):
        if dateiname.endswith('.txt'):
            input_datei_pfad = os.path.join(input_verzeichnis, dateiname)
            output_datei_pfad = os.path.join(output_verzeichnis, f"DEG_{dateiname}")

            # Berechnung durchfÃ¼hren
            berechne_impedanz_und_phase(input_datei_pfad, output_datei_pfad)

# Beispielanwendung
input_verzeichnis = 'D:/Impedanz/ImpedanzGit/impedanz/Sarah_Impedanz/sarah_data_files'  # Pfad zum Verzeichnis mit Input-Dateien
output_verzeichnis = 'D:/Impedanz/ImpedanzGit/impedanz/Sarah_Impedanz/sarah_data_files_Output'  # Pfad zum Verzeichnis fÃ¼r Output-Dateien

verarbeite_alle_dateien(input_verzeichnis, output_verzeichnis)


```

```python
import numpy as np 
import matplotlib.pyplot as plt 

# # Enable LaTeX rendering
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = 'Computer Modern Roman'
plt.rcParams['text.latex.preamble'] = r'\usepackage{amsmath}'

# Globale SchriftgrÃ¶Ãen Ã¤ndern
plt.rcParams.update({
    'axes.titlesize': 20,  # Titel der Achsen
    'axes.labelsize': 18,  # Achsenbeschriftungen
    'xtick.labelsize': 12,  # X-Ticks
    'ytick.labelsize': 12,  # Y-Ticks
    'legend.fontsize': 16,  # Legende
    'font.size': 14         # Allgemeine SchriftgrÃ¶Ãe
})



import matplotlib.pyplot as plt

def zwei_plots(data_x, data_y_one, data_x_schlange, data_y_one_schlange, data_y_two, data_y_two_schlange, name_x, name_y1, name_y2, xscale="", plot_title="title", filename_to_save="LOL.pdf", xmin=0, xmax=500000, ymin1=0, ymax1=1000, ymin2=-2, ymax2=1):
    fig, ax1 = plt.subplots()
    
    ax1.set_xlabel(name_x)
    ax1.set_ylabel(name_y1, color='red')
    
    plot_1 = ax1.plot(data_x, data_y_one, color='red', label='impedance')
    plot_1_schlange = ax1.plot(data_x_schlange, data_y_one_schlange, color='red', label='impedance LT Spice', ls='-.', alpha=0.8) #(alpha=0.2)
    
    ax1.tick_params(axis='y')
    
    if xscale == "log":
        ax1.set_xscale('log') # Setzt die x-Achse auf logarithmisch

    ax1.set_xlim([xmin, xmax])
    ax1.set_ylim([ymin1, ymax1])
    
    ax2 = ax1.twinx()
    ax2.set_ylabel(name_y2, color='green')
    
    plot_2 = ax2.plot(data_x, data_y_two, color='green', label='phase')
    plot_2_schlange = ax2.plot(data_x_schlange, data_y_two_schlange, color='green', label='phase LT Spice', ls='-.', alpha=0.8) #(alpha=0.2)
    
    ax2.tick_params(axis='y')
    
    if xscale == "log":
        ax2.set_xscale('log')  # Setzt die x-Achse auf logarithmisch
    
    ax2.set_xlim([xmin, xmax])
    ax2.set_ylim([ymin2, ymax2])
    
    fig.tight_layout()
    
    lns = plot_1 + plot_2 + plot_1_schlange + plot_2_schlange
    labels = [l.get_label() for l in lns]
    
    plt.title(plot_title)
    plt.legend(lns, labels, loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=2)
    
    #if filename_to_save = "nopdf":
    fnts = filename_to_save + ".pdf"
    plt.savefig(fnts, bbox_inches="tight")
    
    #plt.tight_layout()  # Automatische Anpassung des Layouts
    plt.show()
```
